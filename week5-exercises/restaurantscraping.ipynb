```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("D:\MasterOpleidingDSchijf\DSF\Week 5\pizzeriaVREESWIJK.html") as f: # Original link: http://www.pizzeriavreeswijk.nl/
    	page = BeautifulSoup(f, "lxml")
```

```python
#getting page title
title = page.select("title")[0].get_text()
title
```

```python
# getting info from info blocks
openinghours = page.select(".textwidget")[0].get_text().strip()
contact = page.select(".textwidget")[1].get_text().strip()
delivery =  page.select(".textwidget")[2].get_text().strip()

print(f'Opening hours:\n{openinghours}\n')
print(f'Contact info:\n{contact}\n')
print(f'Delivery info:\n{delivery}\n')
```

```python
# getting social URLS
socials = page.select("#footer-links > li > a")

for social in socials:
   print(social['href'])

# als je available socials wilt zoeken op lijst van websites> attribute selector schrijven om alle http twitter te pakken van een list of websites
```

```python
# changing to menu page
with open("D:\MasterOpleidingDSchijf\DSF\Week 5\pizzas.html", encoding="utf-8") as f:
        	page = BeautifulSoup(f, "lxml")
```

```python
dishes = []
descriptions = []
prices = []

# dishes-location
menuitems = page.select("div.item-text > div.name")

# cleaning and adding it into the list
for menuitem in menuitems:
    menuitem = menuitem.get_text().strip()  
    add = dishes.append(menuitem)

# description-location
menudescriptions = page.select("div.item-text > div.description")

# cleaning and adding it into the list
for menudescription in menudescriptions:
    menudescription = menudescription.get_text().strip()
    add = descriptions.append(menudescription)

# prices-location
menuprices = page.select("div.price")

# cleaning and adding it into list
for menuprice in menuprices: # hier pakt hij die enkele prijs niet
    menuprice = menuprice.get_text().strip()
    add = prices.append(menuprice)
    
#example
print(f'The description of {dishes[0]} is: {descriptions[0]} and costs: {prices[0]}')
```

```python
df = pd.DataFrame()
df["Pizza"] = dishes
df["Description"] = descriptions
df["Price"] = prices

df
```
