```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Use the requests library to get the overview page and feed the HTML of that page to the BeautifulSoup library.
req = requests.get("https://www.voicebooking.com/nl/stemacteur-gezocht/nederlandse-voice-over")
html = req.text

# Converting to local file so I don't ddos them
file = open("vb.html", "w")
file.write(html)
file.close()

# Reopening that same file locally
with open("vb.html") as f:
    page = BeautifulSoup(f, "lxml")
```

```python
dutch_voiceovers = []

#Use the select method of BeautifulSoup to make a list of the items containing the names,
# whether that are table rows, list elements, divs, etcetera.

voiceover_infos = page.select("div.actor-desc > div.actor-name > a")

# Loop through that list and get the basic information out of the rows and append that as a dict to a list.
# incl cleaning
# Also get the link to the user detail page and add that to the list as well.

for voiceover_info in voiceover_infos:
    voiceover_name = voiceover_info.get_text().strip()
    voiceover_href = voiceover_info['href']
    item = {
        "name":voiceover_name,
        "link":voiceover_href
    }
    add = dutch_voiceovers.append(item)
```

```python
# Loop through the list again, 
# fetch the user detail page with requests, parse that using BeautifulSoup and add user detail information to the list.
for item in dutch_voiceovers:
    req = requests.get(item["link"])
    html = req.text
    soup = BeautifulSoup(html, "lxml")
    
    # in-page selections down here - checking for some empty 
    # also printing out what we're getting 
    description = soup.select("div.product.attibute.description > .value > p")
    if not description:
        description = "not_found"
        print(description)
    else:
        description = description[0].get_text().strip()
        print(description)
    
    workedfor = soup.select("div.row.client-s > div.col-md-6 > p")
    if not workedfor:
        workedfor = "not_found"
        print(workedfor)
    else:
        workedfor = workedfor[0].get_text().strip()
        print(workedfor)
    studiogear = soup.select("div.row.client-s > div.col-md-5.col-md-offset-1 > p")
    if not studiogear:
        studiogear: "not_found"
        print(studiogear)
    else:
        studiogear = studiogear[0].get_text().strip()
        print(studiogear)
    checklist = soup.select(".check-list > li")
    language = checklist[0].get_text().strip()
    print(language)
    availability = checklist[1].get_text().strip()
    print(availability)
    contactoptions = checklist[2].get_text().strip()
    print(contactoptions)
    deliverytime = checklist[3].get_text().strip()
    print(deliverytime)
    
    # putting it back into the lists
    item["description"] = description
    item["Worked_For"] = workedfor
    item["Studiogear"] = studiogear
    item["Language"] = language
    item["Availability"] = availability
    item["Contact_Options"] = contactoptions
    item["deliverytime"] = deliverytime
```
